```python
import pandas as pd
import numpy as np
df = pd.read_csv('dataset.csv', header=0)



df = df.assign(avg5day=[0.0]*len(df))
df = df.assign(avg10day=[0.0]*len(df))
df = df.assign(avg30day=[0.0]*len(df))
df = df.assign(D1=[0.0]*len(df))
df = df.assign(D2=[0.0]*len(df))
df = df.assign(D3=[0.0]*len(df))






for i in range(5,len(df)):
    df.at[i,'avg5day'] = df[['Close']].iloc[[i-5,i]].mean(axis=0)
for i in range(10,len(df)):
    df.at[i,'avg10day'] = df[['Close']].iloc[[i-10,i]].mean(axis=0)
for i in range(30,len(df)):
    df.at[i,'avg30day'] = df[['Close']].iloc[[i-30,i]].mean(axis=0)

for i in range(0,len(df)):
    df.at[i,'D1'] = (((df.loc[i,'Close']) - (df.loc[i,'Open']))/(df.loc[i,'Open']))*100000
for i in range(0,len(df)):
    df.at[i,'D2'] = (((df.loc[i,'High']) - (df.loc[i,'Open']))/(df.loc[i,'Open']))*100000
for i in range(0,len(df)):
    df.at[i,'D3'] = (((df.loc[i,'Open']) - (df.loc[i,'Low']))/(df.loc[i,'Open']))*100000

```

```python
dataframe=df[30:732]
```

```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
#from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
# load dataset

dataset = dataframe.loc[:,['D1','Close']].values
# split into input (X) and output (Y) variables
X = dataset[:,0:1].astype(float)

Y = dataset[:,1].astype(float)
# create model
model = Sequential()
model.add(Dense(10, input_dim=1, activation='sigmoid'))
model.add(Dense(1, activation='relu'))
X
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X, Y, epochs=10, batch_size=10)
```

```python
Xnew = numpy.array([[ 12213.481951]])
ynew = model.predict(Xnew)
R=ynew[0][0]
close_value=12213.57
R=R*close_value +close_value
R
```
